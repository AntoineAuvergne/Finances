```python
import pdfplumber

def extract_text_from_pdf(path):
    full_text = ""
    with pdfplumber.open(path) as pdf:
        for page in pdf.pages:
            full_text+=page.extract_text()+ "\n"
    return full_text

#text = extract_text_from_pdf("../../data/raw/Avis-d-opérés-13-11-2025-B-E-SP500-U-ETF-C.pdf")
text = extract_text_from_pdf("../../data/raw/Avis-d-opérés-22-05-2025-ALPHAB-RG-C-NV.pdf")
```
